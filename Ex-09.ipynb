```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = np.zeros((500, 500, 3), dtype=np.uint8)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image1, 'POOJA BRAHMA', (100, 250), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
```

```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
plt.title("Input Image with Text")
plt.axis('off')
```

```python
kernel = np.ones((3, 3), np.uint8)
```

```python
eroded_image = cv2.erode(image1, kernel, iterations=1)
```

```python
plt.imshow(cv2.cvtColor(eroded_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Eroded Image")
plt.axis('off')
```

```python
dilated_image = cv2.dilate(image1, kernel, iterations=1)
```

```python
plt.imshow(cv2.cvtColor(dilated_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title("Dilated Image")
plt.axis('off')
```
